```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "./Data/iris.xlsx"

df = pd.read_excel(file_path)

# Hiển thị thông tin tổng quan về dữ liệu
df.info(), df.head()
```

```python
# Xác định các đặc trưng của Iris
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

# Tạo subplot: mỗi hàng cho một đặc trưng, 2 cột (boxplot và histogram)
fig, axes = plt.subplots(len(features), 2, figsize=(12, 12))

for i, feature in enumerate(features):
    # Vẽ Boxplot
    sns.boxplot(y=df[feature], ax=axes[i, 0], color="lightblue")
    axes[i, 0].set_title(f"Boxplot của {feature}")
    
    # Vẽ Biểu đồ phân phối (histogram với KDE)
    sns.histplot(df[feature], bins=30, kde=True, ax=axes[i, 1], color="lightcoral")
    axes[i, 1].set_title(f"Phân phối của {feature}")

# Điều chỉnh layout để các biểu đồ không chồng chéo
plt.tight_layout()
plt.show()
```
