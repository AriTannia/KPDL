```python
import pandas as pd

# Đọc file CSV vào DataFrame
df = pd.read_csv("C:/Users/ACER/Downloads/bike+sharing+dataset/hour.csv")

num_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

print(df.dtypes)  # Kiểm tra kiểu dữ liệu của từng cột

```

```python
from sklearn.preprocessing import StandardScaler

# Khởi tạo StandardScaler
scaler = StandardScaler()

# Áp dụng chuẩn hóa cho các thuộc tính số
df_scaled = df.copy()
df_scaled[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df_scaled[num_cols].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_scaled[num_cols])
plt.xticks(rotation=45)
plt.title("Boxplot sau khi chuẩn hóa với StandardScaler")
plt.show()
```
