```python
import pandas as pd

# Đọc file CSV vào DataFrame
df = pd.read_csv("C:/Users/ACER/Downloads/bike+sharing+dataset/hour.csv")

num_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

print(df.dtypes)  # Kiểm tra kiểu dữ liệu của từng cột
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Chọn các cột số cần kiểm tra giá trị ngoại lai
num_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

# Vẽ biểu đồ phân phối và boxplot cho từng cột
for col in num_cols:
    plt.figure(figsize=(12, 5))

    # Biểu đồ phân phối (histogram + KDE)
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f"Biểu đồ phân phối của {col}")

    # Boxplot để phát hiện giá trị ngoại lai
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f"Boxplot của {col}")
```
