```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

file_path = "C:/Users/ACER/Downloads/KPDL/Data/Online_Retail_Pre_Processed.xlsx"

df = pd.read_excel(file_path)

# Hiển thị thông tin tổng quan về dữ liệu
df.info(), df.head()
```

```python
features = ['Monetary', 'Frequency', 'Recency']

fig, axes = plt.subplots(len(features), 2, figsize=(12, 12))

for i, feature in enumerate(features):
    # Vẽ Boxplot
    sns.boxplot(y=df[feature], ax=axes[i, 0], color="lightblue")
    axes[i, 0].set_title(f"Boxplot của {feature}")
    
    # Vẽ Biểu đồ phân phối
    sns.histplot(df[feature], bins=30, kde=True, ax=axes[i, 1], color="lightcoral")
    axes[i, 1].set_title(f"Phân phối của {feature}")

plt.tight_layout()
plt.show()
```
